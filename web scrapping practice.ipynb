```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=delhi')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
first_title=soup.find('div', class_="restnt-info cursor")
first_title
```

```python
first_title.text
```

```python
loc=soup.find('div', class_='restnt-loc ellipsis')
loc
```

```python
loc.text
```

```python
price=soup.find('span', class_='double-line-ellipsis')
price
```

```python
price.text.replace('₹', '')
```

```python
titles=[]
for i in soup.find_all('div', class_='restnt-info cursor'):
    titles.append(i.text)
titles
```

```python
location=[]
for i in soup.find_all('div', class_='restnt-loc ellipsis'):
    location.append(i.text)
location
```

```python
price=[]
for i in soup.find_all('span', class_='double-line-ellipsis'):
    price.append(i.text)
price
```

```python
images=[]
for i in soup.find_all('img', class_='no-img'):
    images.append(i['data-src'])
images
```

```python
import pandas as pd
```

```python
print(len(titles), len(location), len(price), len(images))
```

```python
df=pd.DataFrame({'Title':titles, 'Location':location, 'Price':price, 'Image':images})
df
```

```python
#answer 2
page=requests.get('https://www.timeout.com/film/the-100-best-bollywood-movies-the-list')
page
```

```python
top_100_movies=[]
for i in soup.find_all('h3', class_='_h3_cuogz_1'):
    top_100_movies.append(i.text)
top_100_movies
```

```python
df=pd.DataFrame(top_100_movies)[::-1].reset_index(drop=True)
```
